Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
#codes
conditions = [
    (happiness_df['Ladder'] < 30),
    (happiness_df['Ladder'] < 60) & (happiness_df['Ladder'] >= 30),
    (happiness_df['Ladder'] < 90) & (happiness_df['Ladder'] >= 60),
    (happiness_df['Ladder'] >=90) & (happiness_df['Ladder'] <120),
    (happiness_df['Ladder'] >= 120)]
names=["very happy","happy","moderately happy","unhappy","very unhappy"]
happiness_df['happiness_degree'] = np.select(conditions, names)
happiness_df.head(10)
#print(happiness_df.shape)
```

Question 6) Create heat map and comment which values are correlated with each other (15 points)

```python
#codes
#each integer values are correlated
corr=happiness_df.corr()
f, ax = plt.subplots(figsize=(10, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, annot=True)# mask = mask, cmap=cmap
```

Question 9) Plot multiple bar charts in a single chart that show the mean values of 'Social support','Freedom','Generosity' according to happiness degree (15 points)

```python
#codes
w=0.15
df['tmp'] = df.index
values=["Social support","Freedom","Generosity"]
bar1=np.arange(len(df["tmp"]))
bar2=[i+w for i in bar1]
bar3=[j+w for j in bar2]
plt.bar(bar1,df["Social support"],w,label="Social support")
plt.bar(bar2,df["Freedom"],w,label="Freedom")
plt.bar(bar3,df["Generosity"],w,label="Generosity")
plt.xticks(bar1+w,df["tmp"])
del df["tmp"]
plt.legend()
plt.show()
```
